```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_hueristic_df():
    df = pd.read_csv("data/hueristic-df.csv")
    return df 

hueristic_df = read_hueristic_df()

hueristic_df.head()

hueristic_df = hueristic_df[hueristic_df['doc_type'] == 'report']

hueristic_df
```

```python
def read_classifier_df():
    df = pd.read_csv("data/index_reports.csv")
    return df 

classifier_df = read_classifier_df()
classifier_df
```

```python
# precision

# True Positives (TP): 'report' documents in heuristic_df correctly classified as reports
TP = classifier_df[(classifier_df['uid'].isin(hueristic_df[hueristic_df['doc_type'] == 'report']['uid'])) & (classifier_df['score'] >= 0.5)].shape[0]

# False Positives (FP): 'transcript' and 'testimony' documents in heuristic_df incorrectly classified as reports
FP_transcripts_testimonies = classifier_df[(classifier_df['uid'].isin(hueristic_df[(hueristic_df['doc_type'] == 'transcript') | (hueristic_df['doc_type'] == 'testimony')]['uid'])) & (classifier_df['score'] >= 0.5)].shape[0]

# Calculating precision
precision = TP / (TP + FP_transcripts_testimonies)
precision
```

```python
# recall

# True Positives (TP): 'report' documents in heuristic_df correctly classified as reports
TP = classifier_df[(classifier_df['uid'].isin(hueristic_df[hueristic_df['doc_type'] == 'report']['uid'])) & (classifier_df['score'] >= 0.5)].shape[0]

# False Negatives (FN): 'report' documents in heuristic_df NOT classified as reports (scored below 0.5 by the classifier)
FN = classifier_df[(classifier_df['uid'].isin(hueristic_df[hueristic_df['doc_type'] == 'report']['uid'])) & (classifier_df['score'] < 0.5)].shape[0]

# Calculating recall, with a check to avoid division by zero
if TP + FN > 0:
    recall = TP / (TP + FN)
else:
    recall = None # or handle as appropriate for your use case

print(recall)
```

```python
uids_hueristic = set(hueristic_df['uid'])
uids_classifer = set(classifier_df['uid'])

union_uids = uids_hueristic.union(uids_classifer)

intersection_uids = uids_hueristic.intersection(uids_classifer)

unique_uids_hueristic = uids_hueristic - uids_classifer
unique_uids_classifier = uids_classifer - uids_hueristic

count_union_uids = len(union_uids)
count_intersection_uids = len(intersection_uids)
count_unique_uids_hueristic = len(unique_uids_hueristic)
count_unique_uids_classifier = len(unique_uids_classifier)

print("Count of unique uids in the union:", count_union_uids)
print("Count of unique uids in the intersection:", count_intersection_uids)
print("Count of unique uids in classifier_df not in hueristic_df:", count_unique_uids_hueristic)
print("Count of unique uids in classifier_df not in hueristic_df:", count_unique_uids_classifier)
```
